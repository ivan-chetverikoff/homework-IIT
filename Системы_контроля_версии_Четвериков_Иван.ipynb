# Цель работы:
Научиться искользовать систему контроля версий `GIT`, для управления локальным и удалённым репозиторием, работать с ветками, размещать сайты на `GITHub`


### Ходы работы:
1. Установка `GIT` с официального сайта  (https://gitforwindows.org/)
2. Создание локального репозитория
```bash
 mkdir homework-iit
cd homework-iit
git init
```
3. Создание файла отсчёта и фиксация измений
* Создание файла отсчёта `Системы контроля версии_Четвериков_Иван.ipynb`
* Проиндексирование и фиксация изменений
```bash
git add Системы контроля версии_Четвериков_Иван.ipynb
git commit -m "Создан файл с отчётом о проделанных шагах"
```
4. Создание ветки и работа с ней
* Добавление новой ветки
``` bash
git add report.md
git branch feature-branch
git checkout feature-branch
```
* Внесение измений в отчёт и их фиксация
``` bash
git commit -m "Дополнения"
```
5. Слияние изменений в основную ветку
``` bash
git checkout main
git merge feature-branch
```
